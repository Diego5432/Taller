## ¿QUE? 

El artículo nos habla sobre cómo las nuevas tecnologías de información cada vez están mejorando, expandiéndose y normalizando cada vez en el mundo, se centran en los ordenadores, comparando como a través del tiempo se han multiplicado sus existencias conforme a otros países y lo útiles que pueden llegar a ser.


La información recolectada para el estudio nos muestra datos categóricos (Los países) dado que estos solo nos aclaran la región de donde fue sacada la información para el análisis.



Además de esto los datos que nos ofrece el artículo son ordinales, por lo tanto ordenados, ya que la información brindada a pesar de tener una escala de medida donde se compara la cantidad de ordenadores por cada 100 habitantes de cada país, los datos obtenidos no comparten una relación entre sí, la única escala de medida comparativa que está clara son los 100 habitantes, un dato no difiere en el resultado de otro.


Los datos son presentados en un tipo de tabla comparativa, donde la menor cantidad de ordenadores es representada por un tipo de rectángulo claro y entre más oscuro, más cantidad de ordenadores por cada 100 habitantes.

#COVID-19 (EL NUEVO VIRUS)

MAPAAAAA DA.PNG




















#¿QUE?
La visualización de datos que nos ofrece esta información es sobre los casos de covid-19 en todo el mundo en comparación a otros países.
El usuario puede ver que con el pasar del tiempo el virus se va expandiendo y creciendo en comparación a los individuos infectados.
Se puede deducir fácilmente que es un virus desconocido por el impacto que ha causado en tan poco tiempo, muchos infectados, recesión económica, caos y muertes.
Los datos son presentados por un mapa continental que representado por colores y tamaños de puntos representan cuales tienen más contagios y cuáles tienen más cantidad de infectados en cada país. (posiciones) 


#¿POR QUE?
El usuario no necesita hacer transformaciones en los datos, puesto que estos son dados por la oms, mostrándonos el ritmo que toma la pandemia.
Creo que un dato particular que el usuario tiene que tomar en cuenta es el grosor del punto, número acumulado de infectados por país.
Presenta- Identifica- Compara-Localiza

#¿COMO?
En esta visualizacion la escala de color es muy importante, pues presenta que entre mas se aleje del color amarillo se van a ver mas contagios y el grosor del punto muestra el descontrol por la tasa de contagios de covid-19 en el mundo 
El mapa continental muestra al usuario la tendencia del virus y la velocidad con que se expande por todo el mundo.
Se debería usar:
Select -Navigate- Arrange -Filter - Color, luminance, agregar.

#MARCAS
Areas, muestra el espacio territorial de cada país para poder comparar por medio de la tasa de contagios.

#CANAL
Canal, color & shape, por medio del color podemos saber cuántos contagios hay diariamente por cada 100.000 habitantes, y con el tamaño del shape que en este caso es un punto ver como crece la tasa de contagios en el mundo al pasar de los meses.
Canal de posición espacial vertical y horizontal para atributos de ubicación geográfica, Norte, sur, longitud y latitud.  


##RECOMENDACIONES
Creo que la escala de color deberia ser mas clara, tal vez por medio de un degradado sea mas obvio entender en que paises hay mas contagios que en otros y poder ver una tendencia del virus.
